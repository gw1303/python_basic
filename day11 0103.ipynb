# Selenium, Phantomjs

```python
# 웹브라우저를 이용한 스크래핑
# Selenium : 웹 브라우저를 조작(스크래핑) 도구 // 자동으로 url열기, 클릭
# -> 프롬프터 : pip install selenium
# Phantomjs : 화면없이 명령줄에서 이요할 수 있는 웹브라우저
# -> 다운받아서 exe파일을 시스템 환경변수 설정

# Phantomjs를 이용하여 사이트 탐색(이동)
# Selenium을 이용해서 사이트에서 자동으로 url열기, 클릭, 스크롤, 문자입력 가능, 화면캡여 등의 작업을 수행
```

```python
from selenium import webdriver

url = 'www.naver.com'
# Phantomjs 드라이버 추출
browser = webdriver.PhantomJS()

# 3초 대기
browser.implicitly_wait(3)

# url 읽기
browser.get(url)

# 화면캡쳐 및 저장
browser.save_screenshot('myshot.png')

# 브라우저 종료
browser.quit()
```

```python
import requests

url = 'https://stackoverflow.com/questions/tagged/python'
response = requests.get(url)

links = []

soup = BeautifulSoup(response.text,'html.parser')
for link in soup.select('div.question-summary h3 a'):
    links.append(link.attrs['href'])

for link in links:
    url = 'https://stackoverflow.com'+link
    response = requests.get(url)
    text = response.text
    soup = BeautifulSoup(text,'html.parser')
    print('Question\n')
    for x in soup.select('div.postcell div.post-text p'):
        print(x.text)
    print('\nAnswer\n')
    for x in soup.select('div.answercell div.post-text p'):
        print(x.text)
    print('='*40,'\n')
    
#     print(soup)
#     print('Question\n')
    
#     print('\nAnswer\n')
    
#     print('='*50,'\n')
```

```python
# -*- coding: utf-8 -*-
from bs4 import BeautifulSoup
import urllib.request as req

url_list=[]
for i in range(1,101,10):
    url_list.append('https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&c_id=&c_name=&sm=tab_pge&kin_start='+str(i)) 

QT_list=[]
Q_url_list=[]
QA_list=[]
A_list=[]
body =''
# 질문 제목 추출
for url in url_list:
    res = req.urlopen(url)
    soup = BeautifulSoup(res,'html.parser')
    
    for Q in soup.select('dt.question > a'):
        QT_list.append(Q.get_text())
        Q_url_list.append(Q.attrs['href'].replace('§ion','&section'))
        
# 세부 질문 추출        
for url in Q_url_list:
    res = req.urlopen(url)
    soup = BeautifulSoup(res,'html.parser')
    for i in soup.select('div.c-heading__content'):
        QA_list.append(i.get_text())
    for j in soup.select('#SE-ee931118-64a7-40d8-b1f1-208646731bc0 > div > div > div'):
                          #SE-ee931118-64a7-40d8-b1f1-208646731bc0 > div > div > div
                          #SE-57d00d89-6236-4292-b796-ce13aca64af6 > div > div > div
        A_list.append(j.get_text())   
```
