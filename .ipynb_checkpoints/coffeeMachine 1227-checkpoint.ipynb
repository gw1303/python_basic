```python
class CoffeeMachine():
    coffee = 0
    coffeePrice = 200
    money = 0
    balance_total = 0
    balance_100 = 0
    balance_500 = 0
    balance_1000 = 0
    user = 0
    
    def __init__(self):
        print('새로운 자판기가 생성되었습니다.')

    def userChoice(self):
        print('사용자를 선택하세요 \n (1) 관리자 (2) 사용자 (3) 종료')
        self.user = input('입력 : ')
        if self.user == '3' :
            print('안녕히가십시오.')
            self.user = 0
        elif self.user == '2' :
            print('사용자 모드입니다.')
            self.user = 2
            self.userMode()
        elif self.user == '1' :
            print('관리자 모드입니다.')
            self.user = 1
            self.addminMode()
        else :
            print('올바른 사용자를 선택하십시오')

    def addminMode(self):
        print('(1) 커피충전 (2) 잔액  투입 (3) 종료')
        work = input('입력 : ')
        if work == '1':
            self.addCoffee()
        elif work == '2':
            self.addBalacne()
        elif work == '3':
            print('안녕히 가십시오.')
            self.userChoice()

    def addCoffee(self):
        print('현재 커피 : %d잔 \n' %self.coffee)
        add_coffee_num = int(input('커피 충전량을 입력하시오.'))
        if add_coffee_num > 0 :
            self.coffee = self.coffee + add_coffee_num
            print('정상적으로 충전되었습니다. \n 현재 커피 : %d잔' %self.coffee)
            self.addminMode()
        else :
            print('올바른 값을 입력하십시오.')
            self.addCoffee()
    
    def addBalacne(self):
        print('잔액 추가')
        self.balance_100 = self.balance_100 + int(input('100원 동전 몇 개를 투입하시겠습니까?'))
        self.balance_500 = self.balance_500 + int(input('500원 동전 몇 개를 투입하시겠습니까?'))
        self.balance_1000 = self.balance_1000 + int(input('1,000원 지폐 몇 장을 투입하시겠습니까?'))
        self.balance_total = self.balance_total + self.balance_100*100 + self.balance_500*500 + self.balance_1000*1000
        print('''
        현재 잔액
        100원 %d개
        500원 %d개
        1,000원 %d장
        총 %d원 입니다.
        '''%(self.balance_100, self.balance_500, self.balance_1000, self.balance_total))
        self.addminMode()
        
    def userMode(self):
        print('(1) 금액투입 (2) 커피구매 (3) 잔액 반환 (4) 종료', end='\n')
        print('현재 투입금액 : %d원' %self.money, end='\n')
        work = input('입력 : ')
        if work == '1':
            self.addMoney()
        elif work == '2':
            self.buyCoffee()
        elif work == '3':
            self.returnBalance()
        elif work == '4':
            print('안녕히 가십시오.')
            self.userChoice()
    
    def addMoney(self):
        print('금액을 투입하십시오. \n 현재 투입금액 %d원' %self.money)
        money = int(input('입력:'))
        if money > 0:
            self.money = self.money + money
            print('%d원이 투입되어 현재 잔액 %d원 입니다.' %(money, self.money))
            self.userMode()
        else:
            print('올바른 금액을 입력하십시오.')
            self.addMoney()
        
        
    def buyCoffee(self):
        print('커피는 한 잔에 %d원 입니다.' %self.coffeePrice, end='\n')
        print('\n')
        print('현재 커피수량 : %d잔' %self.coffee, end='\n')
        print('현재 투입금액 : %d원' %self.money, end='\n')
        buy_cnt = int(input('몇 잔 주문하시겠습니까?'))
        total_price = buy_cnt * self.coffeePrice
        if buy_cnt > 0 and self.coffee>0 and total_price<=self.money:
            self.coffee = self.coffee - buy_cnt
            self.money = self.money - (self.coffeePrice * buy_cnt)
            print('커피 %d잔이 나왔습니다.' %buy_cnt)
            self.userMode()
        elif self.coffee <= 0:
            print('커피가 부족합니다.')
            self.userMode()
        elif self.coffee < buy_cnt:
            print('구매 수량이 커피 재고보다 많습니다.')
            self.userMode()
        elif self.money < self.coffeePrice or self.money<total_price:
            print('잔액이 부족합니다. 돈을 더 넣어주십시오.')
            self.userMode()
    def returnBalance(self):
        while self.money >= 1000:
            self.balance_1000 -= 1
            self.money -= 1000
            print('1,000원이 나왔습니다.')
            while self.money >= 500:
                self.balance_500 -= 1
                self.money -= 500
                print('500원이 나왔습니다.')
                while self.money != 0:
                    self.balance_100 -= 1
                    self.money -= 100
                    print('100원이 나왔습니다.')
        print('잔액이 반환되었습니다. 금액을 확인하세요.')
        self.userMode()
```

```python
cm1 = CoffeeMachine()
```
