1. A사무실에는 특정일자의 출퇴근 시간이 기록된 거대한 로그파일이 있다고 한다.
파일의 형식은 다음과 같다. (한 라인에서 앞부분은 출근시간(HH:MM:SS), 뒷부분은 퇴근시간이다)
09:12:23 11:14:35
10:34:01 13:23:40
10:34:31 11:20:10
특정시간을 입력(예:11:05:20)으로 주었을 때 그 시간에 총 몇 명이 사무실에 있었는지 알려주는 함수를 작성하시오.

```python
df = pd.read_csv('INOUT.txt',sep=' ')
```

```python
def checkWorker():
    work_time = input('확인 할 시간을 입력하시오 :').replace(':','')
    work_cnt = 0
    
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            df.iloc[i,j]=df.iloc[i,j].replace(':','')

    for i in range(df.shape[0]):
        if df.iloc[i,0] <= work_time <= df.iloc[i,1]:
            work_cnt += 1
    print('%d 명 근무중입니다.' %work_cnt)
    
checkWorker()
```

2.[문제4] DashInsert
DashInsert 함수는 숫자로 구성된 문자열을 입력받은 뒤, 문자열 내에서 홀수가 연속되면 두 수 사이에 - 를 추가하고, 짝수가 연속되면 * 를 추가하는 기능을 갖고 있다. 
(예, 454 => 454, 4546793 => 454*67-9-3)
DashInsert 함수를 완성하자.

입력 - 화면에서 숫자로 된 문자열을 입력받는다.
4546793
출력 - *, -가 적절히 추가된 문자열을 화면에 출력한다.
454*67-9-3

```python
def DashInsert(a):
    res_list=[]
    res=''
    for i in a:
        res_list.append(int(i))
    print(res_list)
    for i in range(len(res_list)-1):
        res+=str(res_list[i])
        if res_list[i]%2 == 0 and res_list[i+1]%2==0:
            res+='*'
        elif res_list[i]%2 == 1 and res_list[i+1]%2==1:
            res+='-'
    res+=str(res_list[-1])
    print(res)
    
DashInsert('4546793')
```

```python
# 3-1
import re
df = pd.read_csv('train.csv')
df
# 호칭 패턴 정의
pat = re.compile('M[ris]+.')
# 호칭 패턴 리스트 생성
M_list=[]

for i in range(df.shape[0]):
    if re.search(pat, df['Name'][i]) != None:
        M_list.append(re.search(pat, df['Name'][i]).group())
print('Mr의 수 : ',M_list.count('Mr.'))
print('Mrs의 수 : ',M_list.count('Mrs.'))
print('Miss의 수 : ',M_list.count('Miss.'))
```
